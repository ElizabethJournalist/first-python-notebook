I analysed Proposition 57- CRIMINAL SENTENCES. JUVENILE CRIMINAL PROCEEDINGS AND SENTENCING

Proposition 57 increased parole chances for felons convicted of nonviolent crimes and gave them more opportunities to earn credits for good behavior. It also allowed judges, not prosecutors, to decide whether to try certain juveniles as adults in court.[2]  (Source: Ballotpedia)


I looked at the open source database and found out that while it had roughfully 60% committees in opposition, Prop 57 received 96% of contributions in support of the proposition.

An interesting fact is that Mark Zuckerberg was the second biggest support contributor of Prop 57.  

The proposition was approved by Californian voters.





```python
import pandas as pd
```

## Imported public domain Commitees' csv file

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

## Imported public domain Contribuitions csv file

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

## Filtered data related to Proposition 57

```python
prop = props[props.prop_name == "PROPOSITION 057 - CRIMINAL SENTENCES. JUVENILE CRIMINAL PROCEEDINGS AND SENTENCING. INITIATIVE CONSTITUTIONAL AMENDMENT AND STATUTE."]
```

## I was not sure if that worked, so I used 'head' to preview the sorted results

```python
prop.head()
```

# I merged filtered proposition series with the 'Contribuiters' csv using the unique identifier 'calaccess_comittee_id' present in both files

```python
merged = pd.merge(prop, contribs, on="calaccess_committee_id")
```

## Again, I used 'head' to verify whether the merge worked or not

```python
merged.head()
```

## Here, I took a look into value counts per committe position.

```python
merged.committee_position.value_counts()
```

## Then, I isolated merged cells for 'Support' and 'Oppose' in order to use Pandas to calculate the expressions below

```python
support = merged[merged.committee_position == "SUPPORT"]
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

## First summed total of 'suport' and oposition' committes 

```python
demco = oppose.committee_position.value_counts()[0] + support.committee_position.value_counts()[0]
```

```python
(float(support.committee_position.value_counts()[0]) / demco) * 100
```

## This time, I summed 'support' contributor amount, then summed 'oppose'

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

## Summed both to find out denominator

```python
dem = support.amount.sum() + oppose.amount.sum()
```

## And now, I could divide the total ammount received by all 'support' committees by the denomintor above to obtain the percent figure

```python
(support.amount.sum() / dem)*100
```

## With the same 'Support' and 'Oppose' breakdowns, I could also look at the major contributers from each side 

```python
support.sort_values("amount", ascending=False).head(10)
```

```python
oppose.sort_values("amount", ascending=False).head(30)
```

## This is just a big picture, to see the top order. It was interesting seeing Mark Zuckerberg showing up as the second biggest supporter of Prop 57. 

```python
merged.groupby(["contributor_firstname","contributor_lastname","committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
